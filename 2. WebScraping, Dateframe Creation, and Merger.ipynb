### Using the first scrape based on title to run a second scrape based on ID, and find the IMDB ID

```python
import time 
counter = 0
tmdb_json2 = []
imdbidlist = []

for id_number in new_list:
    url2 = 'https://api.themoviedb.org/3/movie/' + str(id_number) + '?api_key=e64fe15a762c8a9deb50b7821a5d4356'
    stats = {}
    try:
        response2 = requests.request("GET", url2)
        json_obj2 = response2.json()
        stats['budget'] = json_obj2['budget']
        stats['genres'] = json_obj2['genres']
        stats['revenue'] = json_obj2['revenue']
        stats['runtime'] = json_obj2['runtime']
        stats['title'] = json_obj2['title']
        imdbidlist.append(json_obj2['imdb_id'])
        tmdb_json2.append(stats)
        time.sleep(1)
    except:
        print("{} has no TMDb info".format())
    counter += 1
    if counter % 100 == 0:
        print('* * * Finished {} scrapes * * *'.format(counter))
#     print(soup.find('td', {'class':'scrtext'}).text)
```

```python
imdbdata = []
counter = 0 

for i in imdbidlist:
    url3 = 'https://www.imdb.com/title/'+ i + '/'
    response = requests.get(url3)
    soup = BeautifulSoup(response.content, 'lxml')
    soup.find('div', {'class': 'ratingValue'})
    imdbstats = {}
    try:
        imdbstats["rating"] = soup.find('div', {'class': 'title_wrapper'}).find('div', {'class': 'subtext'}).text[10:-100]
    except: 
        print(imdbstats['title'] + ' has no rating')
    try:
        imdbstats["director"] = soup.find_all('div', {'class': 'credit_summary_item'})[0].find('a').text
    except:
        print(imdbstats['title'] + ' has no director')
    try:
        imdbstats["writer"] = soup.find_all('div', {'class': 'credit_summary_item'})[1].find('a').text
    except:
        print(imdbstats['title'] + ' has no writer')
    try:
        imdbstats["title"] = soup.find('div', {'class': 'title_wrapper'}).find('h1').text[:-8]
    except: 
        Print("NO TITLE FOR THIS MOVIE", i)
    try:
        imdbstats["year"] = soup.find('span', {'id': 'titleYear'}).find('a').text
    except:
        print(imdbstats['title'] + ' has no year')
    try:
        imdbstats["actor_1"] = soup.find_all('div', {'class': 'credit_summary_item'})[2].find_all('a')[0].text
    except: 
        print(imdbstats['title'] + ' has no actor 1')
        
    try: 
        imdbstats["actor_2"] = soup.find_all('div', {'class': 'credit_summary_item'})[2].find_all('a')[1].text
    except: 
        print(imdbstats['title'] + ' has no actor 2')
        
    try:
        imdbstats["actor_3"] = soup.find_all('div', {'class': 'credit_summary_item'})[2].find_all('a')[2].text
    except: 
        print(imdbstats['title'] + ' has no actor 3')
    
    try:
        imdbstats["stars_out_of_10"] = float(soup.find('div', {'class': 'ratingValue'}).text[1:-4])
    except: 
        print(imdbstats['title'] + ' * * * * No Star Rating  * * * * ') 
    
    imdbdata.append(imdbstats)
    
    counter += 1
    if counter % 100 == 0:
        print(' FINISHED {} SCRAPES '.format(counter))
    
    
```
